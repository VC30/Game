```python
import random
print("welcome to the game!")
possible_choice = ["rock","paper","scissors"]
computer_input = random.sample(possible_choice,k=3)
while True:
    player_input = input("make a selection from (rock, paper and scissors): " ).lower()
    if player_input == "rock" or player_input == "paper" or player_input == "scissors":
        print("You selected: "+ player_input + ", it is the computer's turn now...") 
        if player_input == computer_input:
            print("You selected same thing as the computer, it is a tie!")
        elif player_input == "rock":
            if computer_input == "paper":
                print("computer selected paper\npaper wraps rock, computer won!")
            else:
                print("computer selected scissors\nrock smashes scissors, you won!")
        elif player_input == "paper":
            if computer_input == "rock":
                print("computer selected rock\npaper wraps rock, you won!")
            else:
                print("")
                print("computer selected scissors\nscissors cuts paper, computer won!")
        elif player_input == "scissors":
            if computer_input == "rock":
                print("computer selected rock\nrock smashes scissors, computer won!")
            else:
                print("computer selected paper\nscissors cuts paper, you won!")
    else:
        print("You made a wrong selection, try again!!!")
        continue
    to_continue = input("do you want to play again?(y/n): ")
    if to_continue.lower() == "n":
        break
print("Thanks for playing this game!")    
        
```

```python
import random
print("welcome to the game!")
print("make a selection where: 1=rock,2=paper & 3=scissors")
while True:
    user=int(input("input your selection here: "))
    while user > 3 or user < 1:
        user=int(input("invalid input,enter a valid input: "))
    if user == 1:
        user_input = "rock"
    elif user == 2:
        user_input = "paper"
    else:
        user_input = "scissors"
    
    print("Your choice was: " + user_input)
    print("It is computer turn now: .....")
    comp = random.randint(1, 3)
    if comp == 1:
        comp_input = "rock"
    elif comp == 2:
        comp_input = "paper"
    else:
        comp_input = "scissors"
    print("The computer selected: " + comp_input)
    if user_input == comp_input:
        print("You selected same thing as the computer, it is a tie!")
    elif user_input == "rock":
        if comp_input == "paper":
            print("paper wraps rock, computer won!")
        else:
            print("rock smashes scissors, you won!")
    elif user_input == "paper":
        if comp_input == "rock":
            print("paper wraps rock, you won!")
        else:
            print("scissors cuts paper, computer won!")
    elif user_input == "scissors":
        if comp_input == "rock":
            print("rock smashes scissors, computer won!")
        else:
            print("scissors cuts paper, you won!")
    to_continue = input("do you want to play again?(y/n): ").lower()
    if to_continue == "n":
        break
print("Thanks for playing!!!")
```
